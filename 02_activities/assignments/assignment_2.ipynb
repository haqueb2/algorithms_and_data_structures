#### Starter Code for Question 1

```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right
```

```python
from collections import deque

def is_duplicate(root: TreeNode) -> int:
    if root is None:
        return -1 
    
    seen = set()
    queue = deque([root])
    
    while queue: 
        node = queue.popleft() # take from the FRONT (FIFO)
        if node.val in seen:
            return node.val
        else:
            seen.add(node.val)

        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)
            
    return -1
```

```python
# Example 1
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(2)
root.left.left = TreeNode(3)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

is_duplicate(root)
```

```python
# Example 2 
root = TreeNode(1)
root.left = TreeNode(10)
root.right = TreeNode(2)
root.left.left = TreeNode(3)
root.left.right = TreeNode(10)
root.right.left = TreeNode(12)
root.right.right = TreeNode(12)

is_duplicate(root)
```

```python
# Q3 from Assignment 1: 
# If I have a list of integers, I want to find all of the zeros in the list and move them to the end of the list. 
# While doing this, I want to keep the same order of all the non-zero integers in the list. 
# If there are no zeros in the list, I want to return the same list in its original order.
```


-   Create 1 new example that demonstrates you understand the problem. Trace/walkthrough 1 example that your partner made and explain it.


```python
## NEW EXAMPLE ##: 
# Input: 
nums = [0, 1, 0, 3, 12, 0, 5, 0]

# Output: [1, 3, 12, 5, 0, 0, 0, 0]
```

```python
## PARTNER'S EXAMPLE ##:
# Input:
nums = [1, 6, 7, 12, 2]

# Output: [1, 6, 7, 12, 2]

# Explanation: In this example, there are no zeros in the list, so there is nothing to move to the end of the list. Therefore, the output remains the same as the input. 
# A key thing to note is that the order of the non-zero integers is preserved in the output, which is a requirement of the problem.
```


-   Copy the solution your partner wrote. 


```python
from typing import List

def move_zeros_to_end(nums: List[int]) -> List[int]:

    integer_list = []
    zeros_list = []

    for i in nums:
        if i != 0:
            integer_list.append(i)
        else:
            zeros_list.append(i)

    return integer_list + zeros_list

list_1 = [0, 1, 0, 3, 12]
list_2 = [4, 0, 5, 0, 0, 6]
list_3 = [1, 6, 7, 12, 2]
list_4 = [0, 0, 6, 0, 5, 0, 0, 4, 2, 0]

# Applying the user-defined function with our lists.

print(f"Example 1 Output: {move_zeros_to_end(nums=list_1)}")
print(f"Example 2 Output: {move_zeros_to_end(nums=list_2)}")
print(f"New Example 1 Output: {move_zeros_to_end(nums=list_3)}")
print(f"New Example 2 Output: {move_zeros_to_end(nums=list_4)}")
```


-   Explain why their solution works in your own words.


```python
# The solution above works by iterating through each integer in the input list at a time 
# For each integer, it checks if the integer is not equal to zero. 
# If the integer is not equal to zero, it appends the integer to a new list called integer_list. 
# If it is equal to zero, it appends the integer to another new list called zeros_list.
# After iterating through all the integers in the input list, the function concatenates the two lists together, with the integer_list coming first and the zeros_list coming second. 
# This results in a new list that contains all the non-zero integers in their original order, followed by all the zeros in their original order. 
# Finally, the function returns the concatenated list as the output.
```


-   Explain the problem’s time and space complexity in your own words.


```python
# Time complexity: O(n) where n is the length of the input list nums, 
# because you need to iterate through each integer in the list once to check if it is equal to zero or not, and then append it to the appropriate list.
# The amount of time it will take to execute the function is directly proportional to the size of the input list, as each integer must be examined.

# Space complexity: O(n) 
# The function creates two new lists, integer_list and zeros_list, to store the non-zero integers and zeros respectively, 
# the space complexity is O(n) where n is the length of the input list nums. 
# Additionally, it combines these two lists into a new list before returning it, which also contributes to the space complexity.
# The length of all of these lists is proportional to the length of the input list (n), so the overall space complexity is O(n).
```
