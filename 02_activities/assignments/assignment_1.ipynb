
-   Code the solution to your assigned problem in Python (code chunk). Note: each problem can be solved more simply if you use an abstract data type that is suitable for that problem. Using that try to find the best time and space complexity solution!


````python
from typing import List

def first_duplicate(nums: List[int]) -> int:
    seen = set()
    for value in nums:
        if value in seen:
            return value
        else:
            seen.add(value)
    return -1
````

````python
## EXAMPLE 1 ##
# Input: 
nums = [2, 5, 6, 5, 2, 6]

first_duplicate(nums) # Output: 5
````

````python
## EXAMPLE 2 ##
# Input:
nums = [3, 4, 44, 2, 1, 18]

first_duplicate(nums) # Output: -1
````


-   Explain why your solution works


````python
# This function finds the first duplicate value in a list of integers called nums. 
# A set is used to store values it has seen as it iterates through the list.
# For each value in nums, it checks if the value is already in the seen set.
# If it is, that means it's a duplicate and the function returns that value immediately. 
# Since the list if processed from left to right the first duplicate encountered will be the one returned even if there are multiple duplicates in the list.
# If a value is not in the seen set, it adds the value to the set and continues iterating.
# If the loop finishes without finding any duplicates, the function returns -1 to indicate that there are no duplicates in the list.
````


-   Explain the problem’s time and space complexity


````python
# Time complexity: O(n) where n is the length of the input list nums, 
# because in the worst case we may have to check every value in the list to find that there are no duplicates and return -1.
# The use of set allows for O(1) time complexity for the check if a value is already seen and for adding a new value to the set. 
# However, processing all n elements in the list results in O(n) time complexity overall.

# Space complexity: O(n) in the worst case if all elements in nums are unique, because we would have to store all of them in the seen set, which would match the length of the input list.
````
